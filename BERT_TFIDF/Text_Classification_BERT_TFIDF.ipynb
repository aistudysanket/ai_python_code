```python
import numpy as np
import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
path_true_fake ="True_Fake.csv"


# Read both files
df = pd.read_csv(path_true_fake, delimiter=",",
                 engine="python",
                 encoding='utf-8')

```

```python
df.columns
```

```python
df
```

```python
df.label.value_counts()
```

```python
data = df[['text', 'label']]
```

```python
data
```

```python
y = list(data['label'])
```

```python
# y
```

```python
# Word Vectorizing using TFIDF


documents=[]

for i in data['text']:
    s = str(i)
    documents.append(s)
vectorizer = TfidfVectorizer(stop_words='english')
x = vectorizer.fit_transform(documents)
```

```python
documents[0]
```

```python
x[0]
```

```python
# x[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.20,
                                                    random_state=0)
```

```python
lr_clf = LogisticRegression()
lr_clf.fit(x_train, y_train)
```

```python
lr_clf.score(x_test, y_test)
```

```python
lr_clf.predict(x_test)
```

```python
# Word Embeddings using BERT

documents_bert=[]

for i in data['text']:
    s = str(i)
    documents_bert.append(s)

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens') #bert-base-nli-mean-tokens
embeddings = model.encode(documents_bert)
print(embeddings)
```

```python
documents_bert[0]
```

```python
embeddings[0]
```

```python
embeddings.shape
```

```python

train_features, test_features, train_labels, test_labels = train_test_split(embeddings, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(train_features, train_labels)
```

```python
lr_clf.score(test_features, test_labels)
```
