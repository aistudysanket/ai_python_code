## Data Wrangling
- Process of cleaning,organizing and transforming for analysis
- known as Data cleaning

##### Benefits of data wrangling  
- Easy analysis  
- Simple data handling  
- Efficient use of time  
- Clear visualization of data


## Exploratory Data Analysis (EDA)  
- The process of analyzing and summarizing data in order to discover patterns, identify trends, and gain insights.  
- EDA involves using statistical and visualization techniques to understand the characteristics of a dataset and can help inform decisions about how to proceed with further analysis or modeling.

### Comparison of data wrangling and EDA
- Data wrangling and EDA are both concerned with preparing data for analysis, but they differ in terms of their focus and scope.
- Data wrangling is typically more focused on cleaning and organizing data, while EDA is more focused on understanding and summarizing data.
- Data wrangling tends to be more concerned with specific data sets or specific problems, while EDA is more open-ended and can involve looking at multiple data sets or exploring a variety of questions.

```python
# Convert 'sale_date' and 'audit_date' to datetime
df['sale_date'] = pd.to_datetime(df['sale_date'])
df['audit_date'] = pd.to_datetime(df['audit_date'])
```

```python
df
```

```python
df['day'] = df['sale_date'].dt.day
df
```

```python
df['year'] = df['sale_date'].dt.year
df
```

```python
df['month'] = df['sale_date'].dt.month
df
```

```python
df['day_name']=df['sale_date'].dt.day_name()
df
```

```python
#In essence, you're adding a column named 'weekend' that indicates whether a sale occurred on a weekend (True) or a weekday (False).
df['weekend'] = df['sale_date'].dt.day_name().isin(['Saturday', 'Sunday'])
df
```

```python
def weekend_or_weekday(is_weekend):
  if is_weekend:
    return 'Weekend'
  else:
    return 'Weekday'

df['weekend_name'] = df['weekend'].map(weekend_or_weekday)
```

```python
#The map function is used for substituting each value in a Series with another value, derived from a function, a dictionary, or a Series.
df['X'] = df['weekend'].map({True: 'Weekend', False: 'Weekday'})
df
```

```python
# Calculate total weekend revenue
weekend_revenue = df.loc[df['weekend'], 'revenue'].sum()

print(f"Total weekend revenue: {weekend_revenue}")
```

```python
#In essence, the 'Delta' column now contains the duration between the sale date and the audit date for each product.
# This information can be valuable for various analyses, such as understanding the time lag between sales and audits or identifying potential patterns.
df['Delta'] = df['audit_date']-df['sale_date']
df
```

```python
weekend_revenue_sum = df[df['weekend'] == True]['revenue'].sum()
print(weekend_revenue_sum)
```

## The Above code calculates the difference between two dates in days and adds it as a new column called 'diff_days' to your DataFrame, df.

Step-by-step:

   -  `df['Delta']`: This accesses the 'Delta' column of your DataFrame, which presumably contains timedelta values (differences between two dates).
   - `np.timedelta64(1, 'D')`: This creates a NumPy timedelta object representing one day.
    - timedelta is a data type within the datetime module used to represent durations or differences between two points in time. It's essentially a way to store and manipulate spans of time, like the difference between two dates or times.
   - `/`: This operator performs division. It divides each timedelta value in the 'Delta' column by the one-day timedelta. This effectively calculates the number of days in each timedelta.
   - `df['diff_days'] = ...: `This assigns the resulting values (number of days) to a new column called 'diff_days' in your DataFrame.

In simpler terms, you're converting the time differences in the 'Delta' column (which might be in hours, minutes, seconds) into the number of days and storing it in the 'diff_days' column.

----

## Okay, let's break down `np.timedelta64(1, 'D'):`
This code creates a NumPy timedelta object representing a duration of one day.

### Breakdown:

   - np: This refers to the NumPy library, which you've likely imported as import numpy as np.
   - timedelta64: This is a NumPy data type specifically designed for representing time durations or differences.
   - (1, 'D'): These are the arguments passed to timedelta64:
        1: Represents the magnitude or quantity of the duration. In this case, it's 1.
        'D': Specifies the unit of the duration. Here, 'D' stands for days.

In essence, np.timedelta64(1, 'D') creates a timedelta object equivalent to one day.

How it's used:
### New Section
You often use this type of object when performing calculations or comparisons involving dates and times in NumPy or Pandas. For example, you might use it to:

    Add or subtract a specific number of days from a date.
    Calculate the difference between two dates in days.
    Create time-based ranges or sequences.

In the context of your previous code (df['diff_days'] = df['Delta'] / np.timedelta64(1, 'D')), you used it to convert timedelta values (from the 'Delta' column) into the number of days, which were then stored in the 'diff_days' column.
