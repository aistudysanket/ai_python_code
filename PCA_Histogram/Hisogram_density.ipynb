```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
marks= [35, 36, 37, 42, 48, 53, 58, 70, 73, 84]
plt.hist(marks,bins=5,density=True,edgecolor='red')
plt.title("Density Estimation: Marks")
plt.xlabel("Marks")
plt.ylabel("Density")
plt.show()
```

```python
bin_edges = [35, 42, 45, 54, 60, 85 ]
print(bin_edges)
Bin_widths = np.diff(bin_edges)
print(Bin_widths)

plt.hist(marks,bins=bin_edges,density=True,edgecolor='red')
plt.title("Density Estimation: Marks")
plt.xlabel("Marks")
plt.ylabel("Density")
plt.show()
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
alcohol = wine.data[:, 0]  # Alcohol content
malic_acid = wine.data[:, 1]  # Malic acid
```

```python
plt.hist(alcohol, bins=20, density=True)
plt.title("Density Estimation: Alcohol Content")
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(alcohol, rwidth=1.5, density=True)
plt.title("Density Estimation: Alcohol Content")
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.show()
```
